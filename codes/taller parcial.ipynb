\begin{equation}
v(t) = \frac{gm}{c}(1-e^{-\frac{c}{m}t}) .....................     (2)
\end{equation}
- solucion analitica de la aceleracion
\begin{equation}
a(t)=-ge^{\frac{c}{m}t}.....................                (3)
\end{equation}

 # 2
 - derivada de la velocidad por los tres metodos con sus graficas respectivamente

```python

from pylab import plot, show
from math import exp

vela=[]
velc=[]
veld=[]

timea=[]
timec=[]
timed=[]

```

```python
# metodo por  atras
for t in range (1,27):
    
    va=(data[t,1]-data[t-1,1])/(data[t,0]-data[t-1,0]) 
    vela.append(va)
    timea.append(data[t,0])
    
# metodo por el centro    
    
for t in range (1,26):
    
    vc=(data[t+1,1]-data[t-1,1])/(2*(data[t,0]-data[t-1,0]))
    velc.append(vc)
    timec.append(data[t,0])
    
#metodo por adelante

for t in range (1,26):
    
    vd=(data[t+1,1]-data[t,1])/(data[t+1,0]-data[t,0])# hacia adelante
    veld.append(vd)
    timed.append(data[t,0])

plot (timea,vela,'m')
plot (timec,velc,'k')
plot (timed,veld,'c')
show    
```

- calculo de la aceleracion a partir de las velocidades con la velocidad de metodo por atras

```python
acele=[]
timeac=[]
    
for t in range (1,25):
    
    ac=(vela[t]-vela[t-1])/(data[t,0]-data[t-1,0])
    acele.append(ac)
    timeac.append (data[t,0])
    
plot (timeac,acele,'-r')
show    
```

- comparacion entre la solución numerica (usando el método de Euler) y la solucion analitica de la ecuación diferencial teniendo en cuenta una variacion dt
# metodo de euler
Métodos de un paso: Método de Euler, Método de Euler mejorado y Método de Runge-Kutta. En matemática y computación, el método de Euler, llamado así en honor de Leonhard Euler, es un procedimiento de integración numérica para resolver ecuaciones diferenciales ordinarias a partir de un valor inicial dado.

```python

from pylab import plot, show
from math import exp

vana=[]
vnum=[]
vnum2=[]
vnum3=[]

time=[]
tim2=[]
tim3=[]

dt1=1
dt2=1e-1
dt3=1e-3
g=9.8
m=68.1
c=12.5
v1=0
v2=0
v3=0

for t in range (21):
    
    v1=v1+((c/m)*v1-g)*dt1    #velocidad numerica
    v=(m*g/c)*(1-exp(c*t/m)) # velocidad analitica
   
    vana.append (v)
    vnum.append (v1)
    time.append (t)
    
for t in range (201):
    
    v2=v2+((c/m)*v2-g)*dt2    #velocidad numerica
   
    vnum2.append (v2)
    tim2.append (t*dt2)
    
for t in range (20001):
    
    v3=v3+((c/m)*v3-g)*dt3    #velocidad numerica
   
    vnum3.append (v3)
    tim3.append (t*dt3)
    

plot (time,vnum,'b')
plot (time,vana,'g')
plot (tim2,vnum2, 'k')
plot (tim3,vnum3, 'r')

show
```

# metodo de diferenciacion finita
En análisis numérico, el método de las diferencias finitas es utilizado para calcular de manera numérica las soluciones a las ecuaciones diferenciales usando ecuaciones diferenciales finitas para aproximar derivadas.

```python
from pylab import plot, show
from math import exp

Anum=[]
Anum2=[]
Anum3=[]
time1=[]
Aana=[]



for n in range (1,20):
    
    a= -g*exp((c*n)/m) # aceleracion analitica

    dv=(vnum[n]-vnum[n-1])/dt #aceleracion numerica hacia  atras
    
    time1.append (n)
  
    Anum.append (dv)
    
    Aana.append (a)
    
for n in range (1,20):

    dv=(vnum[n+1]-vnum[n])/dt #aceleracion numerica hacia  adelante
  
    Anum2.append (dv)

for n in range (1,20):

    dv=(vnum[n+1]-vnum[n-1])/2*dt #aceleracion numerica centrada
  
    Anum3.append (dv)
    
plot (time1,Aana, 'c')
plot (time1,Anum, 'r')
plot (time1,Anum2,'k')
plot (time1,Anum3,'b')

show
```
